```python
import pandas as pd
import numpy as np
```

```python
books = pd.read_csv('data/Books.csv',  encoding='cp1251', sep=',',on_bad_lines='error')
books.head()
```

```python
# lowercase = books.apply(lambda x: x.str.lower() if x.dtype == 'object' else x)
lowercase = books.apply(
    lambda x: x.str.lower() if x.dtype == 'object' and all(isinstance(val, str) for val in x.dropna()) else x
    )
lowercase.head()
```

```python
ratings = pd.read_csv('data/BX-Book-Ratings.csv',  encoding='cp1251', sep=';',on_bad_lines='skip')
dataset = pd.merge(ratings, lowercase, on=['ISBN'])
dataset.head()
```

```python
# Find all rows where 'Book-Title' is "Siva"
dataset[dataset['Book-Title'].str.lower() == 'siva']
```

```python
# book = 'the fellowship of the ring (the lord of the rings, part 1)'
book = 'siva'
tolkien_readers = dataset['User-ID'][dataset['Book-Title']==book]
tolkien_readers = tolkien_readers.tolist()
tolkien_readers = np.unique(tolkien_readers)

# final dataset
books_of_tolkien_readers = dataset[(dataset['User-ID'].isin(tolkien_readers))]
books_of_tolkien_readers.head()
```

```python
tolkien_readers
```

```python
number_of_rating_per_book = books_of_tolkien_readers.groupby(['Book-Title']).agg('count').reset_index()
number_of_rating_per_book.head()
```

```python
books_to_compare = number_of_rating_per_book['Book-Title'][number_of_rating_per_book['User-ID'] >= 8]
books_to_compare = books_to_compare.tolist()
ratings_data_raw = books_of_tolkien_readers[['User-ID', 'Book-Rating', 'Book-Title']][books_of_tolkien_readers['Book-Title'].isin(books_to_compare)]
ratings_data_raw.head()
```

```python
ratings_data_raw_nodup = ratings_data_raw.groupby(['User-ID', 'Book-Title'])['Book-Rating'].mean()
ratings_data_raw_nodup = ratings_data_raw_nodup.to_frame().reset_index()
ratings_data_raw_nodup
```

```python
dataset_for_corr = ratings_data_raw_nodup.pivot(index='User-ID', columns='Book-Title', values='Book-Rating')
dataset_for_corr.head()
```
